```python
class Square:
    def __init__(self): # __init__   self .
        self.square = int(input('     .\n 1. 2. 3. \n >>>'))

        if self.square == 1:
            print('  rect().')   # Square  rect()  .

        elif self.square == 2:
            print('  par().')   # Square  par()  .

        elif self.square == 3:
            print('  trape().')   # Square  trape()  .

        else:
            print('1, 2, 3   ')

#  

    def par(self):  #    
         width, vertical = map(int, input(',  .  : ,\n >>>').split(','))  #,   
         area = width * vertical  #   
         result = '  : ' + str(area) #   
         return result

a = Square()
a.par()
```

```python
#     !
class Square:
    def __init__(self):
        self.square = int(input('     .\n 1. 2. 3. \n >>>'))

        if self.square == 1:
            print('  rect().')

        elif self.square == 2:
            print('  par().')

        elif self.square == 3:
            print('  trape().')

        else:
            print('1, 2, 3   ')

#  

    def trape(self):
        base1, base2, height = map(int, input(', ,  .  : ,,\n >>>').split(','))
        area = ((base1 + base2) * height) / 2
        result = '  : ' + str(area)
        return result

a = Square()
a.trape()  # 3,2,1     2.5  ..
```

```python
class Kiosk:
    def __init__(self):
        #   
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []  #  
        self.order_price = []  #  
        self.price_sum = 0  #  

    #   
    def menu_print(self):
        print(":")
        for i in range(len(self.menu)):
            print(f"{i + 1}. {self.menu[i]} : {self.price[i]}")

    #  
    def menu_select(self):
        print()  #  
        n = 0
        while n < 1 or len(self.menu) < n:  #    
            n = int(input("   : "))  #   

            if 1 <= n <= len(self.menu):  #    
                self.order_price.append(self.price[n - 1])  #  
                self.price_sum += self.price[n - 1]  #   
            else:  #     
                print(" .   .")

        #   
        t = 0
        while t != 1 and t != 2:  # 1 2   
            t = int(input("HOT  1, ICE  2  : "))
            if t == 1:
                self.temp = "HOT"
            elif t == 2:
                self.temp = "ICE"
            else:
                print("1 2   .\n")

        #   
        self.order_menu.append(self.temp + " " + self.menu[n - 1])
        print(f"{self.temp} {self.menu[n - 1]} : {self.price[n - 1]}")

        #    
        while n != 0:
            print()  #  
            n = int(input("   ,  0  : "))
            if 1 <= n <= len(self.menu):  #   
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT  1, ICE  2  : "))
                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"
                    else:
                        print("1 2   .\n")
                #  
                self.order_menu.append(self.temp + " " + self.menu[n - 1])
                self.order_price.append(self.price[n - 1])
                self.price_sum += self.price[n - 1]
                print(f"   {self.temp} {self.menu[n - 1]} : {self.price[n - 1]}\n : {self.price_sum}")
            elif n == 0:  #   
                print(" .")
                print(f" : {self.order_menu}")
                print(f"  : {self.price_sum}")
            else:
                print(" .   .")

    #  
    def pay(self):
        print(f"   {self.price_sum}.")
        pay_method = 0
        while pay_method != 1 and pay_method != 2:
            pay_method = int(input("  1,   2  : "))
            if pay_method == 1:
                print(" .")
            elif pay_method == 2:
                print("IC    .")
            else:
                print("   .")

    #   
    def table(self):
        # 
        outline1 = "" + "-" * 30 + ""
        outline2 = "|" + " " * 31 + "|"

        #  
        print(outline1)
        for _ in range(5):
            print(outline2)

        #   :  
        for i in range(len(self.order_menu)):
            print(f"{self.order_menu[i]} : {self.order_price[i]}")

        #  
        print(f"  : {self.price_sum}")

        #  
        for _ in range(5):
            print(outline2)
        print(outline1)


# Kiosk   
kiosk = Kiosk()
kiosk.menu_print()
kiosk.menu_select()
kiosk.pay()
kiosk.table()
```
