```python
import re
from collections import Counter

# 파일 읽기
with open('/content/drive/MyDrive/Colab Notebooks/Avengers.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# 소문자로 변환
lower_f = text.lower()

# 기호 제거 (알파벳, 숫자, 공백만 남기기)
cleaned_text = re.sub(r'[^A-Za-z0-9\s]', '', lower_f)

# 단어 단위로 분리
words = cleaned_text.split()

# 각 단어의 빈도수 계산
word_counts = Counter(words)

# 빈도수 순으로 내림차순 정렬
sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# 정렬된 결과에 인덱스 부여
indexed_word_counts = {index: (word, count) for index, (word, count) in enumerate(sorted_word_counts, 1)}

# 결과 출력
print(indexed_word_counts)
```
