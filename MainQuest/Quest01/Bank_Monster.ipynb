Q5. 출금 메서드 : Account 클래스에 출금을 위한 withdraw 메서드를 추가하세요. 출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.

(은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)

Q8. 여러 객체 생성 : Account 클래스로부터 3개 이상 인스턴스를 생성하고 생성된 인스턴스를 리스트에 저장해보세요.

Q9. 객체 순회 반복문을 통해 리스트에 있는 객체를 순회하면서 잔고가 100만원 이상인 고객의 정보만 출력하세요.

Q10. 입금과 출금 내역이 기록되도록 코드를 업데이트 하세요.
(입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드를 추가하세요.)




```python
import random

class Account:
    bank_name = "SC은행"
    account_count = 0  # 클래스 변수: 생성된 계좌 개수

#[계좌개설시] 은행이름, 예금주, 계좌번호 설정
    def __init__(self, owner, balance):   ##_(언더스코어 2개)로 시작하고 끝나는 매직매서드 사용, 객체 생성시에 속성을 초기화 하는 역할.
        self.owner = owner    #예금주이름 ##속성정의 #self: 클래스 안에서 자기 자신(객체)을 참조할 때 사용.
        self.balance = balance #초기잔액
        self.account_number = self.generate_account_number() #계좌번호(랜덤생성)
        self.deposit_count = 0  #입금횟수
        self.deposit_history_list = []      #입출금내역
        self.withdraw_history_list = []
        Account.account_count += 1

#랜덤 계좌번호 생성: 계좌번호는 3자리-2자리-6자리 형태로 랜덤하게 생성
    @staticmethod  #객체와 상관없이 동작하는 메서드 데코레이터
    def generate_account_number():
        return f"{random.randint(100, 999)}-{random.randint(10, 99)}-{random.randint(100000, 999999)}"

#Q3. 클래스 변수 출력 : Account 클래스로부터 생성된 계좌의 개수를 출력하는 get_account_num() 메서드를 추가
    @classmethod   #클래스를 첫번째 인수로 받는 메서드 데코레이터
    def get_account_num(cls):
        return cls.account_count

#Q4. 입금 메서드
    def deposit(self, amount):              ##메서드(동작)
        if amount >= 1:                      #입금금액이 1이상일때만 잔액증가.

            self.balance += amount
            self.deposit_count += 1
            self.deposit_history_list.append(amount)
            if self.deposit_count % 5 == 0:  # 5회 입금마다 1% 이자 추가        #Q7. 이자 지급하기
                self.balance += int(self.balance * 0.01)
#Q5. 출금 메서드
    def withdraw(self, amount):
        if amount <= self.balance:            #출금 금액이 잔액이하일때만 출금.
            self.balance -= amount
            self.withdraw_history_list.append(amount)
#Q6. 정보출력 메서드
    def display_info(self):                   #계좌 정보를 출력
        print(f"은행이름: {self.bank_name}")
        print(f"예금주: {self.owner}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔고: {self.balance:,.0f}원")

#Q10. 입금과 출금내역 기록되도록 코드 업데이트
    def deposit_history(self):                 #입출금 내역 출력
        print("입금 내역:", self.deposit_history_list)

    def withdraw_history(self):
        print("출금 내역:", self.withdraw_history_list)


# Q8: 여러 객체 생성
account1 = Account("파이썬1", 5000)
account2 = Account("파이썬2", 1000000)
account3 = Account("파이썬3", 1500000)
accounts = [account1, account2, account3]

# Q9: 잔고가 100만원 이상인 계좌 정보 출력
for account in accounts:
    if account.balance >= 1000000:
        account.display_info()
print()
# Q10: 입금 및 출금 내역 확인
account1.deposit(10000)                       ##메서드 호출(동작)
account1.withdraw(3000)
account1.deposit(-1000)
account1.deposit(2000)
account1.deposit(50000)
account1.deposit(600000)
account1.deposit(90000)
account1.withdraw(5000000)
account1.deposit_history()
account1.withdraw_history()

balance = account1.balance
deposit_count = 0

for amount in account1.deposit_history_list:
    if amount > 0:  # 정상적인 입금만 반영
        deposit_count += 1
        if deposit_count % 5 == 0:  # 5회 입금 시 이자 발생
            interest = int(balance * 0.01)
            balance += interest
            print(f"{interest}원의 이자가 지급되었습니다.")

# 최종 잔액 출력
print("==============================")
print(f"최종 잔액: {balance:,.0f}원")
```

전투 승리!
고블린 을 이겼다! "



```python
import random

# Q1: Character 클래스 정의
class Character:
    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

#현재 체력이 0 이상인지 bool값을 반환하는 is_alive메서드
    def is_alive(self):
        return self.hp > 0

#공격을 받았을 때, (받은 데미지 - 본인의 방어력)만큼 현재 체력이 감소하는 take_damage 메서드
    def take_damage(self, damage):
        actual_damage = max(0, damage - self.defense)
        self.hp -= actual_damage
        self.hp = max(0, self.hp)

#타겟에게 데미지를 입히는 attack_target 메서드
    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        target.take_damage(damage)
        print(f"{self.name}이(가) {target.name}에게 {damage} 만큼 공격했다 ...!")
        print(f"{target.name}의 체력: {target.hp}\n")

# Q2: Player와 Monster 클래스 정의

#Player클래스-Chracter을 상속받기 레벨1,체력100,공격력25,방어력5로 초기화하기
class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, hp=100, attack=25, defense=5)
        self.experience = 0

#Player클래스-경험치 속성 추가, 인수로 받은 정수 만큼 경험치를 획득하는 gain_experience 메서드 만들기
    def gain_experience(self, exp):
        self.experience += exp
        print(f"{self.name}이(가) 경험치를 {exp} 획득했다!")
        self.level_up()

#현재 경험치가 50이상이면 레벨을 1, 공격력을 10, 방어력을 5씩 올리는 level_up 메서드 만들기
    def level_up(self):
        while self.experience >= 50:
            self.experience -= 50
            self.level += 1
            self.attack += 10
            self.defense += 5
            print(f"{self.name} 레벨업! 현재 레벨: {self.level}, 공격력: {self.attack}, 방어력: {self.defense}\n")

#Monster클래스-Character을 상속받기. 몬스터 레벨에 비례하는 체력,공격력,방어력 초기화하기
class Monster(Character):
    def __init__(self, name, level):
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, hp, attack, defense)

# Q3: battle 함수 정의
def battle(player, monster):
    print(f"{monster.name} 과(와)의 전투를 시작합니다.\n")
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)
        if not player.is_alive():              #살아남지 못한경우 "전투 패배.."출력
            print("전투 패배..\n")
            return False
    print(f"전투 승리! {monster.name} 을(를) 이겼다!\n")
    player.gain_experience(monster.level * 20) #살아남으면 Player 인스턴스에 (몬스터 레벨 * 20)만큼의 경험치를 추가,Player의 레벨업 메서드 호출
    return True

# Q4: main 함수 정의 - 몬스터 이름, 레벨이 매핑된 딕셔너리 정의하기
def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}
    player_name = input("플레이어의 이름을 입력하세요: ") #사용자로부터 이름을 입력받아 player 인스턴스 생성하기.
    player = Player(player_name)

    for monster_name, monster_level in monster_dict.items():
        monster = Monster(monster_name, monster_level)
        if not battle(player, monster):
            print("게임오버")
            break
    else:
        print("축하합니다! 모든 몬스터를 처치했습니다!")

if __name__ == "__main__":
    main()
```
